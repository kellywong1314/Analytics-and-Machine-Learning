# Data Wrangling: American Housing Survey 2015

In this notebook, I read in and examine the survey results from the 2015 American Housing Survey (AHS). The AHS is sponsored by the Department of Housing and Urban Development and conducted by the Census Bureau every odd-numbered year since 1973.

```python
# First, I import pandas and  read in the household.csv file, which contains one entry for each household surveyed. 

import pandas as pd
import numpy as np

df_household = pd.read_csv('Data_Files/household.csv')
```

```python
# examine the column names and values in the first 10 records of the dataframe

pd.set_option('display.max_columns', None)
df_household.head(10)
```

The values of this DataFrame are categoricals that have been encoded. They will need to be translated so that the values can be understood when the analysis is performed. In addition, the column names have been abbreviated. It will be easier to perform analysis if the column names are edited as well.

```python
# use df.info() to get more general information on the DataFrame.

df_household.info()
```

The DataFrame df_household has almost 70,000 entries with 1,091 attributes for each entry. There are only three data types within this DataFrame. For this particular analysis, only some of the attributes will be examined. The DataFrame for the analysis can be reduced to only the relevant columns using "Column Values.xlsx" which has the original names of the columns that will be used for this analysis, and the new column names as well.

```python
# read in an excel file which contains the names of the columns that will be used in this project, and the new names

cols_to_keep = pd.ExcelFile('Data_Files/Column Values.xlsx')
df_colnames = cols_to_keep.parse('new_colnames')
df_colnames.head(10)
```

```python
# use df_colnames to select the columns from df_household that are going to be used for analysis

df_household_clean = df_household[df_colnames['original_col']]
df_household_clean.head()
```

```python
# examine the shape of the new DataFrame, df_household_clean. The DataFrame has only 83 columns now.

df_household_clean.shape
```

```python
# use df.info() to get more general information on the DataFrame. All of the values are stored as int or python objects.

df_household_clean.info(verbose=True)
```

```python
# use .strip() to remove all of the single quotation marks surrounding all of the values.

for column in df_household_clean:
    if df_household_clean[column].values.dtype == 'O':
        df_household_clean[column] = df_household_clean[column].map(lambda x: x.strip("'"))
```

```python
# use .head() to examine the first 5 rows of the DataFrame and confirm no more single quotes remain

df_household_clean.head()
```

```python
# read in "AHS 2015 Value Labels.csv" which has all of the labels for the values in the DataFrame

df_valuelabels = pd.read_csv('Data_Files/AHS 2015 Value Labels.csv', encoding='ISO-8859-1')
df_valuelabels = df_valuelabels.set_index('NAME')
df_valuelabels = df_valuelabels[df_valuelabels.Table == 'HOUSEHOLD']
df_valuelabels.head()
```

```python
# set the index of df_household_clean to the "CONTROL" column

df_household_clean = df_household_clean.set_index('CONTROL')
```

```python
# loop over df_household_clean's columns and map the values from df_valuelabels into the columns using the .map() method

for column in df_household_clean:
    if df_household_clean[column].values.dtype != 'int64' and df_household_clean[column].name != 'OMB13CBSA':
        dict_values = df_valuelabels.loc[column].set_index('Value')['Label'].to_dict()
        df_household_clean[column] = df_household_clean[column].map(dict_values)
        
```

```python
# examine the first five records of df_household_clean to confirm the values have been mapped correctly

df_household_clean.head()
```

```python
# the column OMB13CBSA did not have values in the Value Labels csv, so a dictionary is created to map the metropolitan regions

dict_metro_codes = {'35620':"NY", 
                    '31080':"LA", 
                    '16980':"Chicago",
                    '19100':"Dallas",
                    '37980':"Philadelphia",
                    '26420':"Houston",
                    '47900':"DC",
                    '33100':"Miami",
                    '12060':"Atlanta",
                    '14460':"Boston",
                    '41860':"San Francisco",
                    '19820':"Detroit",
                    '40140':"Riverside",
                    '38060':"Phoenix",
                    '42660':"Seattle",
                    '19740':"Denver",
                    '38300':"Pittsburgh",
                    '38900':"Portland",
                    '17140':"Cincinnati",
                    '17460':"Cleveland",
                    '28140':"Kansas City",
                    '33340':"Milwaukee",
                    '32820':"Memphis",
                    '35380':"New Orleans",
                    '39580':"Raleigh",
                    '99998':"Other Metropolitan",
                    '99999':"Non-Metropolitan"}
```

```python
# use dict_metro_codes to map the correct metropolitan region name to the column OMB13CBSA

df_household_clean['OMB13CBSA'] = df_household_clean['OMB13CBSA'].map(dict_metro_codes)
```

```python
# use the new column names in df_colnames to rename all of the columns in df_household_clean

df_household_clean.columns = df_colnames['new_col'][1:]
df_household_clean.head()
```

```python
# replace the -9 and -6 values in df_household_clean with NaNs

df_household_clean.replace(-9, np.NaN, inplace=True)
df_household_clean.replace(-6, np.NaN, inplace=True)
```

```python
# examine the number of null values in every column

df_household_clean.isnull().sum()
```

```python
# use dropna() to drop any records that have null values in the 'rating_house' and 'rating_neighborhood' columns

df_household_clean = df_household_clean.dropna(subset = ['rating_house', 'rating_neighborhood'])
```

```python
# use dropna() to drop columns with all null values

df_household_clean = df_household_clean.dropna(axis=1, how='all')
```

```python
df_household_clean.isnull().sum()
```

```python
df_household_clean = df_household_clean.drop(['household_income', 'family_income'], axis=1)
```

```python
# the resulting DataFrame has 58,233 entries, 77 columns, and an index column

df_household_clean.info()
```

```python
# save the final dataframe to csv

df_household_clean.to_csv('Data_Files/df_household_clean.csv')
```
